```python
import cv2 as cv
import numpy as np

img=cv.imread("HAND.jpg")
img=cv.resize(img,(1000,600))
img=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
_,ORIGIN=cv.threshold(img,200,255,cv.THRESH_BINARY)
_,thresh=cv.threshold(img,200,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
cv.calcHist(img, [0], None, [256], [0,256])

cv.imshow("gray_image",img)
cv.imshow("thresh_image",ORIGIN)
cv.imshow("otsu_image",thresh)


cv.waitKey(0)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
  
img = cv2.imread('sanjeev.jpg',0)

# create a mask
mask = np.zeros(img.shape[:2], np.uint8)
mask[100:3000, 2500:4000] = 255
masked_img = cv2.bitwise_and(img,img,mask = mask)
masked_img=cv2.resize(masked_img,(1000,600))
cv2.imshow("mask",masked_img)

# Calculate histogram with mask and without mask
hist_full = cv2.calcHist([img],[0],None,[256],[0,256])
hist_mask = cv2.calcHist([img],[0],mask,[256],[0,256])

plt.subplot(221), plt.imshow(img, 'gray')
plt.subplot(222), plt.imshow(mask,'gray')
plt.subplot(223), plt.imshow(masked_img, 'gray')
plt.subplot(224), plt.plot(hist_full), plt.plot(hist_mask)
plt.xlim([0,256])

plt.show()
cv2.waitKey(0)
```
