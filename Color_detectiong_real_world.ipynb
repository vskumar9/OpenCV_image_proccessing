```python
import cv2 as cv
import numpy as np

#capturing video through webcam
cap=cv.VideoCapture(0)

while(1):
    _,img=cap.read()
    
    #converting frame(img==bgr) to HSV(hue-saturationn-value)
    hsv=cv.cvtColor(img,cv.COLOR_BGR2HSV)
    
    #red color
    red_lower=np.array([136,87,111],np.uint8)
    red_upper=np.array([180,255,255],np.uint8)
    
    #blue color
    blue_lower=np.array([99,80,2],np.uint8)
    blue_upper=np.array([120,255,255],np.uint8)
    
    #yellow color
    yellow_lower=np.array([22,180,200],np.uint8)
    yellow_upper=np.array([60,255,255],np.uint8)
    
    #white color
    white_lower=np.array([0,0,200],np.uint8)
    white_upper=np.array([180,20,255],np.uint8)
    
    #black color
    black_lower=np.array([0,0,0],np.uint8)
    black_upper=np.array([180,255,30],np.uint8)
    
    #green color
    green_lower=np.array([25,52,72],np.uint8)
    green_upper=np.array([102,255,255],np.uint8)
    
    #all color together
    red=cv.inRange(hsv,red_lower,red_upper)
    blue=cv.inRange(hsv,blue_lower,blue_upper)
    yellow=cv.inRange(hsv,yellow_lower,yellow_upper)
    white=cv.inRange(hsv,white_lower,white_upper)
    black=cv.inRange(hsv,black_lower,black_upper)
    green=cv.inRange(hsv,green_lower,green_upper)
    
    #Morphological Transform,Dilation
    
    kernal=np.ones((5,5),"uint8")
    
    red=cv.dilate(red,kernal)
    res_red=cv.bitwise_and(img,img,mask=red)
    
    blue=cv.dilate(blue,kernal)
    res_blue=cv.bitwise_and(img,img,mask=blue)
 
    yellow=cv.dilate(yellow,kernal)
    res_yellow=cv.bitwise_and(img,img,mask=yellow)
    
    white=cv.dilate(white,kernal)
    res_white=cv.bitwise_and(img,img,mask=white)
    
    black=cv.dilate(black,kernal)
    res_black=cv.bitwise_and(img,img,mask=black)
    
    green=cv.dilate(green,kernal)
    res_green=cv.bitwise_and(img,img,mask=green)
    
    
    #tracking red
    contours,hierarchy=cv.findContours(red,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)
    for pic, contour in enumerate(contours):
        area=cv.contourArea(contour)
        if(area>300):
            x,y,w,h=cv.boundingRect(contour)
            img=cv.rectangle(img,(x,y),(x+w,y+w),(0,0,255),2)
            cv.putText(img,"Red Color",(x,y),cv.FONT_HERSHEY_SIMPLEX,1,(0,0,255))
    #Tracking blue
    contours,hierarchy=cv.findContours(blue,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)
    for pic, contour in enumerate(contours):
        area=cv.contourArea(contour)
        if(area>300):
            x,y,w,h=cv.boundingRect(contour)
            img=cv.rectangle(img,(x,y),(x+w,y+w),(255,0,0),2)
            cv.putText(img,"Blue Color",(x,y),cv.FONT_HERSHEY_SIMPLEX,1,(255,0,0))
    #Tracking green
    contours,hierarchy=cv.findContours(green,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)
    for pic, contour in enumerate(contours):
        area=cv.contourArea(contour)
        if(area>300):
            x,y,w,h=cv.boundingRect(contour)
            img=cv.rectangle(img,(x,y),(x+w,y+w),(0,255,0),2)
            cv.putText(img,"Green Color",(x,y),cv.FONT_HERSHEY_SIMPLEX,1,(0,255,0))
    #Tracking yellow
    contours,hierarchy=cv.findContours(yellow,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)
    for pic, contour in enumerate(contours):
        area=cv.contourArea(contour)
        if(area>300):
            x,y,w,h=cv.boundingRect(contour)
            img=cv.rectangle(img,(x,y),(x+w,y+w),(0,255,0),2)
            cv.putText(img,"Yellow Color",(x,y),cv.FONT_HERSHEY_SIMPLEX,1,(0,255,0))
    #Tracking white
    contours,hierarchy=cv.findContours(white,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)
    for pic, contour in enumerate(contours):
        area=cv.contourArea(contour)
        if(area>300):
            x,y,w,h=cv.boundingRect(contour)
            img=cv.rectangle(img,(x,y),(x+w,y+w),(255,255,255),2)
            cv.putText(img,"White Color",(x,y),cv.FONT_HERSHEY_SIMPLEX,1,(255,255,255))
    #Tracking black
    contours,hierarchy=cv.findContours(black,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE)
    for pic, contour in enumerate(contours):
        area=cv.contourArea(contour)
        if(area>300):
            x,y,w,h=cv.boundingRect(contour)
            img=cv.rectangle(img,(x,y),(x+w,y+w),(0,0,0),2)
            cv.putText(img,"Red Color",(x,y),cv.FONT_HERSHEY_SIMPLEX,1,(0,0,0))
    
    cv.imshow("Color Tracking",img)
    if cv.waitKey(10)&0xFF==ord('q'):
        cap.release()
        cv.destroyAllWindows()
        break
```
